```python
import pandas as pd

yield_rates1 = pd.read_csv('data/bond-yield/yield-curve-rates-1990-2021.csv')
yield_rates1
```

```python
spread_10_3 = []

for i in range(len(yield_rates1)):
    spread_10_3.append(yield_rates1.loc[i]['10 Yr'] - yield_rates1.loc[i]['3 Mo'])

# spread_10_3
```

```python
dates = pd.to_datetime(yield_rates1['Date'])

df_reversed = pd.DataFrame({
    'Date': dates,
    '10 Yr': yield_rates1['10 Yr'],
    '3 Mo': yield_rates1['3 Mo'],
    'Spread': spread_10_3
})

df = df_reversed.copy().dropna().iloc[::-1]
df = df.set_index(df['Date'])
df = df.drop(columns='Date')
df_2005_09 = df['2005':'2009']
df_2005_09.to_csv('2005_2009_10_yr.csv')
```
